### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
vacation_df = cities_df.loc[(cities_df['Max Temp']>75),:]
vacation_df = vacation_df.loc[(cities_df['Max Temp']<90),:]
vacation_df = vacation_df.loc[(cities_df['Humidity']<50),:]
vacation_df = vacation_df.loc[(cities_df['Cloudiness']<40),:]
vacation_df = vacation_df.loc[(cities_df['Wind Speed']<9),:]
vacation_df = vacation_df.drop(columns=['Unnamed: 0'])
vacation_df = vacation_df.drop_duplicates(subset='City')
vacation_df = vacation_df.reset_index(drop=True)
vacation_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
vacation_df['Hotel Name'] = ''
hotel_df = vacation_df.copy()
hotel_df['Lat'] = hotel_df['Lat'].map('{:,.2f}'.format)
hotel_df['Lng'] = hotel_df['Lng'].map('{:,.2f}'.format)
hotel_df
```

```python
# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
  
# use iterrows to iterate through pandas dataframe
for index, row in hotel_df.iterrows():
    
    try:
        # get cooridinates from df
        lat = row['Lat']
        lng = row['Lng']
        target_coordinates = f'{lat}, {lng}'

        # set radius and type parameters
        target_radius = 5000
        target_type = "lodging"

        # set up a parameters dictionary
        params = {
        "location": target_coordinates,
        "radius": target_radius,
        "type": target_type,
        "key": g_key
        }

        # assemble url and make API request
        print(f"Retrieving Results for Index {index}: {target_coordinates}.")
        response = requests.get(base_url, params=params).json()

        # add information to dataframe
        hotel_df.loc[index,'Hotel Name'] = response['results'][0]['name']
        
    except:
#         hotel_df.loc[index,'Hotel Name'] = ''
        print('**Missing field, skipping to next query.')
    
```

```python
hotel_df2 = hotel_df.copy() 
hotel_df2 = hotel_df2.loc[(hotel_df2['Hotel Name'] != ''),:]
hotel_df2
```

```python
hotel_df2 = hotel_df2.reset_index(drop=True)
hotel_df2
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df2.iterrows()]
locations = hotel_df2[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
cooridinates = []

# Assign the marker layer to a variable
for index, row in hotel_df2.iterrows():
    
        # get cooridinates from df
        lat = row['Lat']
        lng = row['Lng']
        cooridinates.append((float(lat), float(lng)))
cooridinates        
```

```python
markers = gmaps.marker_layer(cooridinates,hotel_info)

# Add the layer to the map
fig.add_layer(markers)

# Display figure
fig
```
